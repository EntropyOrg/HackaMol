###Example of using the HackaMol::Molecule Class
First load HackaMol and Modern::Perl

```perl
use Modern::Perl;
use HackaMol;
use Time::HiRes qw(time);
```

The text file corresponding to pdbid 2CBA is downloaded into the same directory as this notebook. 

```perl
`ls *pdb`
```

####Tasks: 
1. Load 2CBA.pdb into an instance of HackaMol::Molecule
2. Calculate the center of mass
3. Calculate the radius of gyration 
4. Translate to recenter the molecule
5. recalculate COM and Rg
6. time it all

```perl
my $t1 = time;

my $mol = HackaMol->new
                  ->read_file_mol("2CBA.pdb"); # object methods chain from left to right

my $t2 = time;

say $mol->COM;
say $mol->Rg;
$mol->translate(-$mol->COM);
say $mol->COM;
say $mol->Rg;

my $t3 = time;

printf ("Time to load: %10.3fs, Time to do some calculations: %10.3f\n", $t2-$t1, $t3-$t2);
```

```perl
$mol->print_pdb;
```

####Strip out the backbone and measure dihedral angles

```perl
my @N_CA_C = grep { $_->name eq 'N'  or 
                    $_->name eq 'CA' or 
                    $_->name eq 'C' 
             } $mol->all_atoms;

my @dihedrals = HackaMol->new
                        ->build_dihedrals(@N_CA_C);

foreach my $dihe (@dihedrals) {
    printf( "%-25s ", $dihe->name );
    foreach my $t (0 .. $mol->tmax){
      $dihe->gt($t);
      printf( "%7.2f ", $dihe->dihe_deg );
    }
    print "\n";
}
```

### Print out backbone xyz to screen

```perl
HackaMol::Molecule->new (
                        atoms=>[@N_CA_C]
                    )->print_xyz;
```

### Print out backbone pdb to file

```perl
HackaMol::Molecule->new (
                        atoms=>[@N_CA_C]
                    )->print_pdb("2CBA_bb.pdb");
```
